```python
def calculate_distance(lat1, lon1, lat2, lon2):
    """Haversine distance calculation"""
    import math
    
    # Handle NaN values
    if any(pd.isna([lat1, lon1, lat2, lon2])):
        return 0
    
    # Convert to radians
    lat1, lon1, lat2, lon2 = map(math.radians, [lat1, lon1, lat2, lon2])
    
    # Haversine formula
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = (math.sin(dlat/2)**2 + 
         math.cos(lat1) * math.cos(lat2) * math.sin(dlon/2)**2)
    c = 2 * math.asin(math.sqrt(a))
    
    return 6371 * c  # Distance in km

df_merged['Latitude_next'] = df_merged['Latitude'].shift(-1)
df_merged['Longitude_next'] = df_merged['Longitude'].shift(-1)

df_merged['distance_to_next'] = df_merged.apply(
    lambda row: calculate_distance(
        row['Latitude'], row['Longitude'],
        row['Latitude_next'], row['Longitude_next']
    ), axis=1
)
```

```python


# --- CORRECTED Feature Lists (based on your DataFrame) ---
# RobustScaler for outlier-prone features
robust_features = ['Latitude', 'Longitude', 'Speed', 'distance_to_next', 'Accuracy', 'acceleration']

# StandardScaler for rolling statistics
standard_features = ['speed_rolling_mean_5', 'speed_rolling_std_5']

# MinMaxScaler for bounded features
minmax_features = ['day_of_week', 'day_of_year', 'cumulative_time_minutes']

# --- Get a list of the columns that are actually in the DataFrame ---
available_cols = list(df_merged.columns)

# Filter each feature list to keep only the available columns
# This step makes the code robust, even if a column is missing in the future.
robust_features_present = [col for col in robust_features if col in available_cols]
standard_features_present = [col for col in standard_features if col in available_cols]
minmax_features_present = [col for col in minmax_features if col in available_cols]

# Combine all features that will be transformed
all_present_features = set(robust_features_present + standard_features_present + minmax_features_present)


# --- Create the ColumnTransformer with the valid column lists ---
preprocessor = ColumnTransformer(
    transformers=[
        ('robust', RobustScaler(), robust_features_present),
        ('standard', StandardScaler(), standard_features_present),
        ('minmax', MinMaxScaler(), minmax_features_present)
    ],
    remainder='passthrough'  # Keep all other columns untouched
)

# Apply the transformations
df_merged_scaled_array = preprocessor.fit_transform(df_merged)

# --- Convert back to a DataFrame with correct column names ---
# Find columns that were not transformed
passthrough_columns = [col for col in df_merged.columns if col not in all_present_features]

# The new column order is the one defined by the ColumnTransformer
new_column_order = robust_features_present + standard_features_present + minmax_features_present + passthrough_columns

# Create the new scaled DataFrame
df_merged_scaled = pd.DataFrame(df_merged_scaled_array, columns=new_column_order, index=df_merged.index)

print("Scaling complete!")
# You can now use the `df_merged_scaled` DataFrame
print(df_merged_scaled.head())
```
